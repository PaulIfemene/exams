#PART A
We are comparing the average crime rate in Uganda (46.6) to towns where the crime rate is small (Group 1, with crime rates of less than or equal to 0.7).

```python
from scipy.stats import ttest_1samp

# Subset the data into Group 1 and Group 2
group1 = df[df['CRIM'] <= 0.7]['CRIM'].dropna()

group2 = df[df['CRIM'] > 0.7]['CRIM'].dropna()
#Defining the mean of uganda
uganda_mean = 46.6

# Perform one-sample t-tests
t_stat_uganda, p_value_uganda = ttest_1samp(group1, uganda_mean)

# Summary statistics for Group 1
group1_stats = group1.describe()
group1_stats, t_stat_uganda, p_value_uganda
```

 We found that the average crime rate in Uganda = 46.6 is way higher than the average crime rate of the small-crime towns (which is only 0.16) which means that Uganda has more crimes than these other towns.

We’re comparing the average crime rate in the United States (49.2) to towns with bigger crime rates (Group 2, where the crime rate is more than 0.7).



```python
# Defining mean for US
us_mean = 49.2

# Perform one-sample t-tests
t_stat_us, p_value_us = ttest_1samp(group2, us_mean)

# Summary statistics for group 2
group2_stats = group2.describe()
group2_stats, t_stat_us, p_value_us
```
